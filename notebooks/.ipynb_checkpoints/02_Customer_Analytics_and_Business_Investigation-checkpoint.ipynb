# Customer Analytics & Business Investigation

- Do customers purchasing across multiple product categories generate greater value?
- Which customer purchasing patterns contribute the highest profitability?
- Does the first product category purchased influence future purchasing behaviour?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option("display.max_columns", None)
pd.set_option("display.float_format", "{:.2f}".format)
```

# Load Customer-Level Analytical Dataset

The customer-level dataset created in Notebook 1 is loaded for advanced customer analysis.

```python
df = pd.read_csv(
    "../data/Sample_Superstore.csv",
    encoding="latin1"
)

df["Order Date"] = pd.to_datetime(df["Order Date"])

customer_summary = pd.read_csv(
    "../data/customer_summary.csv",
    index_col="Customer ID"
)

customer_summary.head()
```

# Validate the Customer-Level Dataset

Before beginning the investigation, the analytical dataset is validated to ensure that it has been loaded correctly and contains the customer-level features required for the subsequent analysis.

The analytical dataset has been successfully transformed from a transaction-level dataset into a customer-level dataset.

# Business Investigation 1

## Business Question

The exploratory analysis indicated that Furniture generated relatively lower profitability than Technology.

However, category-level summaries cannot determine whether Furniture customers themselves are low-value customers.

Many customers purchase products across multiple categories, meaning their overall contribution to the business may be significantly greater than Furniture profitability alone suggests.

To investigate this, customer profit is separated by product category.

```python
# ==========================================================
# Category-wise Customer Profit
# ==========================================================

furniture_profit = (
    df[df["Category"] == "Furniture"]
    .groupby("Customer ID")["Profit"]
    .sum()
    .rename("Furniture_Profit")
)

technology_profit = (
    df[df["Category"] == "Technology"]
    .groupby("Customer ID")["Profit"]
    .sum()
    .rename("Technology_Profit")
)

office_profit = (
    df[df["Category"] == "Office Supplies"]
    .groupby("Customer ID")["Profit"]
    .sum()
    .rename("Office_Supplies_Profit")
)

customer_summary = customer_summary.join(
    [furniture_profit, technology_profit, office_profit]
).fillna(0)

customer_summary.head(10)
```

### Analyst Interpretation

Customer profitability cannot be fully understood through a single profit metric. By separating profit into **Furniture**, **Technology**, and **Office Supplies**, the analysis captures **how each customer generates value across different product categories**.

This customer-centric view enables deeper investigation into purchasing patterns, cross-category behavior, and category-specific profitability. Rather than asking *how profitable a customer is*, the analysis can now determine *where that profitability comes from*.

These engineered features provide the analytical foundation for identifying high-value customers, evaluating category dependence, and developing targeted business strategies throughout the remainder of the project.

# Business Investigation 2

## Business Question

If customers purchase products across multiple categories, do they generate greater business value than customers purchasing from only one category?

Understanding this relationship may reveal opportunities for cross-selling and customer retention strategies.

To answer this question, a new feature is created to measure the number of product categories in which each customer generated positive profit.

```python
customer_summary["Multi_Category"] = (
    (customer_summary["Furniture_Profit"] > 0).astype(int)
    + (customer_summary["Technology_Profit"] > 0).astype(int)
    + (customer_summary["Office_Supplies_Profit"] > 0).astype(int)
)

customer_summary.head(15)
```

### Why This Matters

The **Multi_Category** feature measures the **breadth of each customer's purchasing behavior** by counting the number of product categories in which the customer generated **positive profit**. Rather than focusing on transaction volume, this feature captures how diversified a customer's purchasing relationship is with the business.

Customers who purchase profitably across multiple product categories may represent **higher-value and more loyal customers**, as they are likely to have broader engagement with the company's product portfolio. In contrast, customers purchasing from only one category may present opportunities for **cross-selling and revenue expansion**.

The next analysis investigates whether customers with broader purchasing behavior generate **higher sales, greater profitability, and stronger long-term business value**, providing insights that can support customer retention, product bundling, and cross-selling strategies.

# Business Investigation 3

## Business Question

The previous investigation created a measure of customer purchase breadth using the **Multi_Category** feature.

The next question is whether customers purchasing across multiple product categories generate greater business value than customers purchasing from only one category.

If customers purchasing from multiple categories consistently generate higher profitability, this may indicate opportunities for cross-selling and stronger customer relationship strategies.

The following analysis compares customer profitability, category contribution, and purchasing behaviour across different levels of category participation.

```python
profile_analysis = (
    customer_summary.groupby("Multi_Category")
    .agg({
        "Total_Profit": ["count", "mean", "median"],
        "Furniture_Profit": "mean",
        "Technology_Profit": "mean",
        "Office_Supplies_Profit": "mean",
        "Total_Orders": "mean"
    })
    .round(2)
)

profile_analysis
```

## Analyst Interpretation

The **Multi_Category** feature provides a way to evaluate whether **customer purchase breadth** is associated with greater business value.

Rather than measuring customers solely by the number of orders they place, this analysis compares profitability, order frequency, and category-level contributions across customers who purchase profitably from different numbers of product categories.

If customers purchasing across multiple categories consistently generate **higher profits and place more orders**, it would suggest that encouraging **cross-category purchasing** could improve customer lifetime value and strengthen long-term profitability.

However, knowing **how many** categories customers purchase from is only part of the story. The next analysis identifies the **specific category combinations** purchased by customers to determine which purchasing patterns create the greatest business value and where cross-selling opportunities exist.

# Business Investigation 4

## Business Question

Customers purchasing from two categories may behave very differently depending on which categories they purchase.

For example:

- Furniture + Technology
- Furniture + Office Supplies
- All Three Categories

Although these customers all purchase multiple categories, their profitability and purchasing behaviour may differ substantially.

To investigate these differences, customers are grouped according to the categories from which they generated positive profit.

```python
customer_summary["has_furniture"] = (
    customer_summary["Furniture_Profit"] > 0
).astype(int)

customer_summary["has_technology"] = (
    customer_summary["Technology_Profit"] > 0
).astype(int)

customer_summary["has_office"] = (
    customer_summary["Office_Supplies_Profit"] > 0
).astype(int)
```

### Creating Customer Groups

The following function classifies customers according to the combination of product categories from which they generated positive profit.

This produces customer groups that can be compared throughout the remainder of the investigation.

```python
def categorize_customer(row):

    if (
        row["has_furniture"] == 1
        and row["has_technology"] == 0
        and row["has_office"] == 0
    ):
        return "Furniture Only"

    elif (
        row["has_furniture"] == 1
        and row["has_technology"] == 1
        and row["has_office"] == 0
    ):
        return "Furniture + Technology"

    elif (
        row["has_furniture"] == 1
        and row["has_technology"] == 0
        and row["has_office"] == 1
    ):
        return "Furniture + Office Supplies"

    elif (
        row["has_furniture"] == 1
        and row["has_technology"] == 1
        and row["has_office"] == 1
    ):
        return "All Three"

    else:
        return "Other"
```

```python
customer_summary["Customer_Group"] = customer_summary.apply(
    categorize_customer,
    axis=1
)

customer_summary["Customer_Group"].value_counts()
```

## Analyst Observation

Grouping customers by the **combination of product categories** they purchase provides a deeper understanding of purchasing behaviour than analyzing individual product categories in isolation.

Rather than treating all Furniture customers as a single group, this approach distinguishes between customers who purchase **only Furniture** and those who also purchase **Technology** and/or **Office Supplies**. This allows the analysis to capture differences in customer value, purchasing breadth, and profitability.

By shifting the focus from products to **customer purchasing patterns**, the analysis establishes a stronger foundation for identifying **cross-selling opportunities**, evaluating **customer profitability**, and developing **targeted retention strategies**.

## Comparing Customer Groups

After identifying distinct customer purchasing patterns, the next step is to determine **which groups create the greatest business value**.

This analysis compares customers based on the combination of product categories from which they generate positive profit, allowing differences in profitability and purchasing behavior to be evaluated across groups.

The comparison includes the following metrics:

- **Number of customers**
- **Average customer profit**
- **Median customer profit**
- **Average Furniture profit**
- **Average Technology profit**
- **Average Office Supplies profit**
- **Average number of orders**

Together, these metrics provide a comprehensive view of customer value, enabling the identification of the most profitable purchasing patterns and uncovering opportunities for **cross-selling**, **customer retention**, and **profitability optimization**.

```python
group_analysis = (
    customer_summary[
        customer_summary["Customer_Group"] != "Other"
    ]
    .groupby("Customer_Group")
    .agg({
        "Total_Profit": ["count", "mean", "median"],
        "Furniture_Profit": "mean",
        "Technology_Profit": "mean",
        "Office_Supplies_Profit": "mean"
    })
    .round(2)
)

group_analysis.columns = [
    "Count",
    "Mean_Profit",
    "Median_Profit",
    "Furniture",
    "Technology",
    "Office_Supplies"
]

group_analysis
```

## Analyst Interpretation

The comparison reveals that **customer purchasing breadth is strongly associated with profitability**.

Customers who generate profit across **all three product categories** deliver the **highest average and median profit**, while customers purchasing **Furniture only** generate the **lowest profitability**, with a negative average profit.

This finding suggests that **customer value should be evaluated holistically rather than by individual product categories alone**. Encouraging customers to purchase across multiple categories may increase customer lifetime value, improve profitability, and reduce dependence on low-margin product purchases.

The next analysis investigates **customer purchasing journeys** to understand how these high-value customer relationships develop and identify opportunities to encourage similar purchasing behavior.

# Business Investigation 5

## Business Question

Customer groups have been identified based on their purchasing behaviour.

The next question is whether these customer groups also differ in **how their relationship with the business begins**.

Understanding the first product category purchased by customers can provide valuable insights into customer acquisition and future purchasing behaviour.

Although the dataset cannot establish causality, identifying common entry points may help the business understand which categories attract long-term valuable customers.

```python
# Sort transactions chronologically and identify
# the first category purchased by each customer

first_category_purchased = (
    df.sort_values("Order Date")
      .groupby("Customer ID")["Category"]
      .first()
)
```

```python
# Map the first purchased category
# to the customer-level dataset

customer_summary["First_Category"] = (
    customer_summary.index.map(first_category_purchased)
)
```

```python
first_cat_by_group = (
    customer_summary[
        customer_summary["Customer_Group"] != "Other"
    ]
    .groupby("Customer_Group")["First_Category"]
    .value_counts()
)

first_cat_by_group
```

## Analyst Interpretation

This analysis examines whether **high-value customer groups tend to begin their purchasing journey with different product categories**.

By identifying the **first product category** purchased by each customer, the analysis provides insight into potential customer acquisition patterns and how purchasing relationships evolve over time.

Although the results do not establish a cause-and-effect relationship, they can reveal whether certain product categories are more effective at attracting customers who later become **high-value, multi-category purchasers**.

These insights can support **customer acquisition**, **cross-selling**, and **marketing strategies** by identifying which product categories are most likely to serve as entry points for long-term profitable customer relationships.

# Business Investigation 6

## Business Question

Furniture profitability varies across customer groups.

Rather than evaluating Furniture performance at an overall category level, the next investigation measures how much Furniture profit is generated by each customer group.

This helps determine whether certain customer groups contribute disproportionately to Furniture profitability.

```python
furniture_profit_by_group = (
    customer_summary[
        customer_summary["Customer_Group"] != "Other"
    ]
    .groupby("Customer_Group")
    .agg(
        Total_Furniture_Profit=("Furniture_Profit", "sum"),
        Customer_Count=("Furniture_Profit", "count"),
        Avg_Profit_Per_Customer=("Furniture_Profit", "mean")
    )
)

furniture_profit_by_group
```

```python
current_furniture_profit = (
    furniture_profit_by_group[
        "Total_Furniture_Profit"
    ].sum()
)

print(
    f"Current Total Furniture Profit: "
    f"{current_furniture_profit:.2f}"
)
```

## Analyst Interpretation

The analysis reveals that **Furniture profitability is not evenly distributed across customer groups**.

Customers purchasing across **all three product categories** contribute the **largest share of total Furniture profit**, while customers purchasing **Furniture and Technology** generate the **highest average Furniture profit per customer**.

In contrast, **Furniture-only customers** contribute the **lowest total and average Furniture profit**, indicating that purchasing Furniture in isolation is associated with significantly lower customer value.

These findings suggest that **cross-category purchasing strengthens Furniture profitability**. Rather than focusing solely on acquiring more Furniture customers, the business should encourage existing Furniture customers to purchase complementary products through **cross-selling strategies**, thereby increasing customer lifetime value and overall profitability.

# Business Investigation 7

## Business Question

Customer retention is often more cost-effective than customer acquisition.

To understand the potential business risk associated with losing different customer groups, a hypothetical churn scenario is created.

**Important Note**

The Superstore dataset does **not** contain actual churn labels.

Therefore, the following analysis is **not** a predictive churn model.

Instead, it estimates the potential financial impact under different hypothetical churn assumptions.

```python
# Hypothetical churn assumptions

churn = {
    "Furniture Only": 0.90,
    "Furniture + Technology": 0.30,
    "Furniture + Office Supplies": 0.20,
    "All Three": 0.05
}
```

```python
impact = furniture_profit_by_group.copy()

impact["Churn_Rate"] = (
    impact.index.map(churn)
)
```

```python
impact["Customers_Lost"] = (
    impact["Customer_Count"]
    * impact["Churn_Rate"]
).round()
```

```python
impact["Profit_Lost_From_Churn"] = (
    impact["Customers_Lost"]
    * impact["Avg_Profit_Per_Customer"]
)
```

```python
impact["Furniture_Profit_Lost"] = (
    impact["Total_Furniture_Profit"]
)
```

```python
impact["Total_Impact"] = (
    impact["Furniture_Profit_Lost"]
    + impact["Profit_Lost_From_Churn"]
)

impact
```

```python
total_loss = impact["Total_Impact"].sum()

print(
    f"Estimated Total Business Impact: "
    f"{total_loss:.2f}"
)
```

## Analyst Interpretation

Customer churn should be evaluated based on its **financial impact**, not simply the number of customers lost.

The analysis demonstrates that customer groups contribute differently to overall business value. Although some groups experience higher estimated churn, the greatest financial risk arises when **high-value, multi-category customers** are lost because they generate substantially greater profit over time.

This approach enables the business to prioritize customer retention efforts according to **expected financial impact** rather than churn rates alone. Targeting the retention of profitable multi-category customers is likely to produce a greater return on investment than applying the same retention strategy to all customer groups equally.

Overall, the estimated business impact highlights the importance of combining **customer profitability**, **purchase behavior**, and **churn risk** when making strategic customer retention decisions.

# Business Investigation 8

## Business Question

One possible explanation for Furniture's relatively low profitability is the use of discounts.

While discounts may increase sales volume, excessive discounting can also reduce overall profitability.

To investigate this relationship, Furniture customers are grouped according to their average discount level.

The objective is to understand whether customers receiving higher discounts also generate higher long-term business value.

```python
# Select customers generating positive Furniture profit

furniture_buyers = customer_summary[
    customer_summary["Furniture_Profit"] > 0
].copy()
```

```python
# Create discount buckets

discount_segments = pd.cut(
    furniture_buyers["Avg_Discount"],
    bins=[0,0.05,0.10,0.15,1.0],
    labels=[
        "0-5%",
        "5-10%",
        "10-15%",
        "15%+"
    ],
    include_lowest=True
)
```

```python
discount_analysis = (
    furniture_buyers
    .groupby(discount_segments)
    .agg(
        Customer_Count=("Total_Profit","count"),
        Mean_Profit=("Total_Profit","mean"),
        Median_Profit=("Total_Profit","median"),
        Furniture_Profit=("Furniture_Profit","mean"),
        Technology_Profit=("Technology_Profit","mean"),
        Office_Supplies_Profit=("Office_Supplies_Profit","mean"),
        Avg_Orders=("Total_Orders","mean")
    )
    .round(2)
)

discount_analysis
```

## Analyst Interpretation

The analysis indicates a **clear relationship between higher discount levels and lower customer profitability**.

Customers receiving **low to moderate discounts (0–10%)** generate the highest average profit, while profitability declines as discount levels increase. Customers receiving **discounts above 15%** produce the lowest average customer profit and substantially lower profits across Furniture, Technology, and Office Supplies.

These findings suggest that **aggressive discounting may reduce overall business profitability**, particularly when discounts exceed moderate levels. While discounts may help drive sales volume, excessive discounting appears to erode profit margins without delivering proportionate customer value.

However, this analysis identifies an **association rather than a causal relationship**. The dataset does not include product costs, competitive pricing, promotional objectives, or customer acquisition strategies, all of which may influence discount decisions. Additional business context would be required before recommending changes to the current pricing strategy.

# Appendix: Customer Health Analysis

## Why this analysis?

The primary objective of this project was to investigate the profitability of the Furniture category.

After answering the main business question, an additional customer health analysis was conducted to understand how customer acquisition timing relates to long-term purchasing behaviour.

This section is exploratory and complements the main business investigation.

## Customer Cohort Analysis

### Business Question

Do customers acquired in different years exhibit different purchasing behaviour and profitability?

To answer this, customers were grouped according to the year of their first purchase (cohort year), allowing comparison across acquisition periods.

```python
print(df.columns.tolist())

# First purchase date
first_order = (
    df.groupby("Customer ID")["Order Date"]
      .min()
)

customer_summary["Acquisition_Date"] = (
    customer_summary.index.map(first_order)
)

# Last purchase date
last_order = (
    df.groupby("Customer ID")["Order Date"]
      .max()
)

customer_summary["Last_Order_Date"] = (
    customer_summary.index.map(last_order)
)

# Customer lifetime (months)
customer_summary["Months_Since_Acquisition"] = (
    (
        customer_summary["Last_Order_Date"]
        - customer_summary["Acquisition_Date"]
    ).dt.days / 30
)

# Cohort year
customer_summary["Cohort_Year"] = (
    customer_summary["Acquisition_Date"].dt.year
)
```

```python
cohort_analysis = (
    customer_summary
    .groupby("Cohort_Year")
    .agg(
        Customer_Count=("Total_Profit","count"),
        Mean_Profit=("Total_Profit","mean"),
        Median_Profit=("Total_Profit","median"),
        Avg_Orders=("Total_Orders","mean"),
        Avg_Discount=("Avg_Discount","mean")
    )
    .round(2)
)

cohort_analysis
```

### Insight

The cohort analysis provides a high-level view of customer quality across acquisition years.

Rather than evaluating individual customers, this analysis compares the average purchasing behaviour of customers acquired during different periods.

## Customer Retention After One Year

### Business Question

How do customers who remain active for at least one year compare across different acquisition cohorts?

Filtering customers with at least twelve months of observed activity reduces the influence of recently acquired customers and enables a fairer comparison of long-term purchasing behaviour.

```python
# Customers with at least one year of activity
year_one = customer_summary[
    customer_summary["Months_Since_Acquisition"] >= 12
].copy()

cohort_year_one = (
    year_one
    .groupby("Cohort_Year")
    .agg(
        Mean_Profit=("Total_Profit","mean"),
        Mean_Orders=("Total_Orders","mean")
    )
    .round(2)
)

cohort_year_one
```

### Insight

Customers with at least one year of observed activity continue to generate meaningful purchasing behaviour across acquisition cohorts.

Since the dataset represents only a limited observation period rather than complete customer lifetimes, these results should be interpreted as indicators of customer engagement rather than definitive lifetime value.
